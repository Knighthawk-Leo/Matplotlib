```python
import matplotlib.pyplot as plt
#importing the required package

```

```python
x=[1,2,3,4,5]
y=[0,5,10,15,20]
plt.plot(x,y)
plt.show()
```

```python
#Now  we change the color in three ways
#by writing name of color
plt.plot(x,y, color='yellow')
plt.show()
```

```python
#using hex-code
plt.plot(x,y, color='#008000')
plt.show()
```

```python
#using short key-word
plt.plot(x,y, color='m') #“m”= magenta
plt.show()
```
